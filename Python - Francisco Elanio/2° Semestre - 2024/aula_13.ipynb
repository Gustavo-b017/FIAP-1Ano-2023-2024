```python
# merge e concat de dados

left = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

# Merge dos dados

df_merge = pd.merge(left, right, on='key')
df_merge
```

```python
# Concatenando dados

df_concat = pd.concat([left, right])
df_concat
```

```python
left = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    'A': ['C0', 'C1', 'C2', 'C3'],
    'B': ['D0', 'D1', 'D2', 'D3']
})

df_concat = pd.concat([left, right], axis = 0).reset_index(drop=True)
df_concat
```
